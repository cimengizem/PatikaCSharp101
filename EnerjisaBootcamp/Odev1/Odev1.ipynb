### **Outliers of Radius Mean**

```python
# separate data in similar situation 
df_bening = df[df["diagnosis"] == "B"]
df_malignant = df[df["diagnosis"] == "M"]
```

```python
# view the values of variable radius_mean and detect outliers of radius_mean
desc = df_bening.radius_mean.describe()
Q1 = desc[4]
Q3 = desc[6]
IQR = Q3 - Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
print("Anything outside this range is an outlier: (", lower_bound, "," , upper_bound, ")")
df_bening[df_bening.radius_mean < lower_bound].radius_mean
print("Outliers: ", df_bening[(df_bening.radius_mean < lower_bound) | (df_bening.radius_mean > upper_bound)].radius_mean.values)
```

* Mean ve median değerleri arasında çok büyük farklılıklar gözlenmemektedir. Normal dağılıma yakınsak verilere sahip olduğumuz söylenebilir.
* lower_bound olan 7.645000000000001 değerinin altında 1 outlierımız, upper_bound olan 16.805 değerimizin üzerinde 2 outlier değerimiz bulunmaktadır.
* Outlier değerlerimizi incelediğimiz zaman çok uç değerlere sahip olmadığını görüyoruz.

## Soru 3

```python
df_corr = df.loc[:,['radius_mean','area_mean','compactness_mean']]
df_corr
```

### Correletion

```python
# correletion
df_corr.corr()
```

* Korelasyon matrisine göre radius_mean ile area_mean arasında yüksek pozitif korelasyon ilişkisi vardır. 
* Korelasyon matrisine göre radius_mean ile compactness_mean arasında düşük pozitif korelasyon vardır.
* Korelasyon matrisine göre area_mean ile compactness_mean arasında düşük pozitif korelasyon vardır.

### Covariance

```python
# covariance
df_corr.cov()
```

* Kovaryans incelemelerinde 3 değerinde birbiri ile ilişkileri pozitif yönde değişim göstermektedir. 

### Pearson Correlation Coefficient

```python
# pearson correlation coefficient
df_corr.corr(method = 'pearson')
```

* Pearson korelasyon katsayısı metodu ile hesaplanan değerler korelasyon matrisi ile hesaplanan değerler ile benzer sonuç vermiştir.
* Pearson korelasyon matrisine göre radius_mean ile area_mean arasında yüksek pozitif korelasyon ilişkisi vardır. 
* Pearson korelasyon matrisine göre radius_mean ile compactness_mean arasında düşük pozitif korelasyon vardır.
* Pearson korelasyon matrisine göre area_mean ile compactness_mean arasında düşük pozitif korelasyon vardır.

### Spearman Rank Coefficient

```python
# spearman rank coefficient
df_corr.corr(method = 'spearman')
```

* Spearman rank katsayısı metodu ile hesaplanan değerler, korelasyon matrisi ile hesaplanan değerler ile ufak farklılıklar göstermiştir.
* Spearman rank katsayısına göre radius_mean ile area_mean arasında diğer yöntemlere göre daha yüksek pozitif korelasyon ilişkisi gözlemlenmiştir. 
* Spearman rank katsayısına göre radius_mean ile compactness_mean arasında diğer yöntemlere göre küçük bir oranda düşüş gözlenmiştir ve düşük pozitif korelasyon vardır.
* Spearman rank katsayısına göre area_mean ile compactness_mean arasında diğer yöntemlere göre küçük bir oranda düşüş gözlenmiştir ve düşük pozitif korelasyon vardır.

### Effect Size

```python
# effect size
# eftype --> cohen / Cohen d effect size
df_rac = pg.compute_effsize(df_corr.radius_mean, df_corr.area_mean, paired = False, eftype = "cohen")
df_rcc = pg.compute_effsize(df_corr.radius_mean, df_corr.compactness_mean, paired = False, eftype = "cohen")
df_acc = pg.compute_effsize(df_corr.area_mean, df_corr.compactness_mean, paired = False, eftype = "cohen")
# eftype --> hedges / Hedges g effect size 
df_rah = pg.compute_effsize(df_corr.radius_mean, df_corr.area_mean, paired = True, eftype = "hedges")
df_rch = pg.compute_effsize(df_corr.radius_mean, df_corr.compactness_mean, paired = True, eftype = "hedges")
df_ach = pg.compute_effsize(df_corr.area_mean, df_corr.compactness_mean, paired = True, eftype = "hedges")
print("Cohen d Effect Size")
print("-Radius Mean and Area Mean with Cohen d: ",df_rac)
print("-Radius Mean and Compactness Mean: ",df_rcc)
print("-Area Mean and Compactness Mean: ",df_acc)
print("Hedges g Effect Size")
print("-Radius Mean and Area Mean with Hedges g: ",df_rah)
print("-Radius Mean and Compactness Mean: ",df_rch)
print("-Area Mean and Compactness Mean: ",df_ach)
```

* Verilen değerlerin Effect Size değerlerine ulaşmak için iki farklı yöntem kullanılmıştır.
* Cohen d ve Hedges g birbirine çok benzer yöntemlerdir. 
* Her iki yöntem ile hesaplanan effect size değerleri birbirlerine çok yakın sonuçlar vermiştir. 
* Cohen d yönteminde d değeri 0.2'den küçük ise düşük, 0.2 ve 0.8 arasında ise orta ve 0.8'den büyük ise yüksek büyüklükte etkiye sahiptir.
* Cohen d effect size yönteminde d değeri 0.8'den büyük olduğu için karşılaştırılan her değer için etki büyüklüğü kuvvetli yorumunu yapabiliriz. Bu değerlendirme sonucunda her bir verinin beraber kullanımlarının tek kullanılmalarına kıyasla daha etkili sonuçlar vereceğini söyleyebiliriz.
* Sonuçlarımızı karşılaştırdığımızda ise en büyük etkinin Radius Mean ve Compactness Mean arasında olduğu gözlenmektedir. Bu durumda bu iki verinin beraber kullanılması en etkili sonuçları verecektir.
